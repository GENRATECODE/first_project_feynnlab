```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("new features.csv")
```

```python
data.head(10)
```

```python
# encoding labels
from sklearn.preprocessing import LabelEncoder
label_encoding=LabelEncoder()
data["label_encoded"]=label_encoding.fit_transform(data["Activity"])
data
```

```python
data1=data.drop("Activity",axis=1)
```

```python
# spliting test and train data 
x=data1.iloc[:,:-1]
y=data1.iloc[:,-1:]
```

```python
#From sklearn.model_selection import train_test_split
# train and test data 
from sklearn.model_selection import train_test_split
Train_x,Test_x,Train_y,Test_y=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# knn fit 
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
Knn_fit=knn.fit(Train_x,Train_y)
```

```python
Knn_predict=Knn_fit.predict(Test_x.values)
```

```python
# Here model is prepared without hyperparameter tuning 
# check accuracy
from sklearn import metrics 
accuracy=metrics.accuracy_score(Test_y,Knn_predict)
print("accuracy: {:.4f}".format(accuracy))
```

```python
# hyperparameter tuning for knn
# manual tuning 
Knn_pro=KNeighborsClassifier(n_neighbors=1,weights="distance",leaf_size=50)
```

```python
knn_fit1=Knn_pro.fit(Train_x,Train_y)
```

```python
Knn_pre=knn_fit1.predict(Test_x.values)
```

```python
from sklearn import metrics
print("Accuracy:{:.4f}".format(metrics.accuracy_score(Knn_pre,Test_y)))
```
