# Hybrid model(knn+svm)

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("new_Features.csv")
```

```python
data
```

```python
# label Encoding 
from sklearn.preprocessing import LabelEncoder
label_encode= LabelEncoder()
data["Activity_code"]=label_encode.fit_transform(data["Activity"])
data1=data.drop("Activity",axis=1)  # Removing the activity colum with label and assigning in new variable
```

```python
data["Activity_code"].value_counts()
```

```python
# Splitting the data into x, y 
x=data1.iloc[:,:-1]
y=data1.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
Train_X,Test_X,Train_Y,Test_Y=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# buliding one model 
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
estimators = []
model1=SVC(C=1.0, kernel='poly', degree=4,tol=0.001)
estimators.append(("SVC",model1))

model2=KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski')
estimators.append(("KNN",model2))

ensemble=VotingClassifier(estimators)
```

```python
hybrid=ensemble.fit(Train_X,np.ravel(Train_Y,order="c")) # np.ravel is used to convert multi row  array into 1d with nothing chnged
```

```python
hybrid_pred=hybrid.predict(Test_X.values)
```

```python
from sklearn import metrics
print("Accuracy : {:.4f}".format(metrics.accuracy_score(Test_Y,hybrid_pred)))
```

```python
from sklearn.metrics import confusion_matrix,f1_score,recall_score,precision_score
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(Test_Y,hybrid_pred)
report = classification_report(Test_Y,hybrid_pred)
confusion = confusion_matrix(Test_Y,hybrid_pred)

print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')
print(f'Confusion Matrix:\n{confusion}')
```
